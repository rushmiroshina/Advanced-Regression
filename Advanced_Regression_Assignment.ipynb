### Data Preparation

```python
#creating dummies
categorical = df.select_dtypes(include=['category'])
dummy = pd.get_dummies(categorical)
```

```python
df = df.drop(list(categorical.columns),axis=1)
df.head(3)
```

```python
df = pd.concat([df, dummy], axis=1)
df.head()
```

### Setting up the data for creating the model

```python
df_train,df_test = train_test_split(df,train_size=0.7,test_size=0.3,random_state=42)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
#Scaling dataset
scaler = StandardScaler()
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
# checking distribution of target variable
plt.figure(figsize=(15,7))
plt.subplot(121)
sns.distplot(df_train.SalePrice)
plt.subplot(122)
sns.distplot(df_test.SalePrice)
```

```python
# x & y split
y_train = df_train.pop('SalePrice')
X_train = df_train

y_test = df_test.pop('SalePrice')
X_test = df_test
```

### Lasso Regression

```python
lasso = Lasso(alpha=0.001)
lasso.fit(X_train,y_train)

y_train_pred = lasso.predict(X_train)
print(r2_score(y_true=y_train,y_pred=y_train_pred))

y_test_pred  = lasso.predict(X_test)
print(r2_score(y_true=y_test,y_pred=y_test_pred))
```

```python
#using gridSearchCV to find the best alpha value

folds = KFold(n_splits=10,shuffle=True,random_state=42)

hyper_param = {'alpha':[0.0001,0.001, 0.01, 0.1,1.0, 5.0,7.0,9.0,10.0,15.0,17.0,20.0]}

model = Lasso()

model_cv = GridSearchCV(estimator = model,
                        param_grid=hyper_param,
                        scoring='r2',
                        cv=folds,
                        verbose=1,
                        return_train_score=True
                       )

model_cv.fit(X_train,y_train)
```

```python
lasso_result = pd.DataFrame(model_cv.cv_results_)
lasso_result.head()
```

```python
plt.figure(figsize=(16,8))
plt.plot(lasso_result['param_alpha'],lasso_result['mean_train_score'],color ='blue',label='train')
plt.plot(lasso_result['param_alpha'],lasso_result['mean_test_score'],color ='red',label='test')
plt.xscale('log')
plt.ylabel('R2 Score')
plt.xlabel('Alpha')
plt.legend()
plt.show()
```

```python
# best parameter-Alpha value
model_cv.best_params_
```

```python
lasso = Lasso(alpha=0.001)
lasso.fit(X_train,y_train)

y_train_pred = lasso.predict(X_train)
y_test_pred = lasso.predict(X_test)

print(r2_score(y_true=y_train,y_pred=y_train_pred))
print(r2_score(y_true=y_test,y_pred=y_test_pred))
```

```python
model_param = list(lasso.coef_)
model_param.insert(0,lasso.intercept_)
cols = df_train.columns
cols.insert(0,'const')
lasso_coef = pd.DataFrame(list(zip(cols,model_param)))
lasso_coef.columns = ['Feature','Coef']
```

```python
lasso_coef.sort_values(by='Coef',ascending=False).head(10)
```

#### Neighborhood_NridgHt,Neighborhood_Timber,Exterior1st_AsbShng,BsmtQual_Fa,BsmtFullBath 
are the top features obtained for Lasso Regression 

### Ridge Regression

```python
ridge = Ridge(alpha=0.001)
ridge.fit(X_train,y_train)

y_train_pred = ridge.predict(X_train)
print(r2_score(y_train,y_train_pred))
y_test_pred = ridge.predict(X_test)
print(r2_score(y_test,y_test_pred))
```

```python
folds  = KFold(n_splits=10,shuffle=True,random_state=42)

hyper_param = {'alpha':[0.001,0.01,0.1,0.2,0.5,0.9,1.0, 5.0, 10.0,20.0,30.0,40.0]}

model = Ridge()

model_cv = GridSearchCV(estimator=model,
                        param_grid=hyper_param,
                        scoring='r2',
                        cv=folds,
                        verbose=1,
                        return_train_score=True)

model_cv.fit(X_train,y_train)
```

```python
ridge_result = pd.DataFrame(model_cv.cv_results_)
ridge_result.head()
```

```python
plt.figure(figsize=(16,8))
plt.plot(ridge_result['param_alpha'],ridge_result['mean_train_score'],color ='blue',label='train')
plt.plot(ridge_result['param_alpha'],ridge_result['mean_test_score'],color ='green',label='test')
plt.xlabel('Alpha')

plt.ylabel('R2 Score')
plt.legend()
plt.show()
```

```python
# best parameter-Alpha value
model_cv.best_params_
```

```python
ridge = Ridge(alpha = 20)
ridge.fit(X_train,y_train)

y_pred_train = ridge.predict(X_train)
print(r2_score(y_train,y_pred_train))

y_pred_test = ridge.predict(X_test)
print(r2_score(y_test,y_pred_test))
```

```python
model_parameter = list(ridge.coef_)
model_parameter.insert(0,ridge.intercept_)
cols = df_train.columns
cols.insert(0,'constant')
ridge_coef = pd.DataFrame(list(zip(cols,model_parameter)))
ridge_coef.columns = ['Feature','Coef']
```

```python
ridge_coef.sort_values(by='Coef',ascending=False).head(10)
```

#### Neighborhood_NridgHt,BsmtQual_Fa,Neighborhood_Timber,OverallCond,KitchenQual_Fa
are the top features for Ridge Regression
